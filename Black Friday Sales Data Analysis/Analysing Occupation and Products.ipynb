```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('D:/Data_Analysis/BlackFriday.csv')

del df['Product_Category_2']
del df['Product_Category_3']
```

```python
df.head()
```

```python
sns.countplot(x = df['Stay_In_Current_City_Years'], hue = df['Gender'])
```

```python
df.groupby('Stay_In_Current_City_Years').size().plot(kind = 'bar')
```

```python
sns.countplot(x = df['Occupation'])
```

# ->Which Occupation is purchasing what no. of products ?

```python
df.groupby('Occupation').size().sort_values().plot(kind = 'bar', figsize = (8,4))
```

# ->Which Occupation has more purchasing power ?

```python
df.groupby('Occupation')['Purchase'].sum().sort_values().plot(kind = 'bar')
```

# ->Which Occupation is buying more expensive products ?

```python
df.groupby("Occupation")['Purchase'].mean().sort_values()
```

```python
df.groupby("Occupation")['Purchase'].mean().sort_values().plot(kind = 'bar')
```

```python
sns.countplot(x = df['Occupation'], hue = df['Marital_Status'])
```

# ->Unique_Products buying from each Occupation category

```python
df.groupby('Occupation')['Product_ID'].nunique().plot(kind = 'bar')
```

```python
df.groupby("Occupation")['Product_ID'].nunique().sort_values().plot(kind = 'bar')
```

# ->No. of products in each category

```python
df.groupby('Product_Category_1').size().sort_values().plot(kind = 'bar')
```

# ->Amount spent on each category

```python
df.groupby('Product_Category_1').sum()['Purchase'].sort_values().plot(kind = 'bar')
```

# ->Which product in Category-1 is more costly ?

```python
df.groupby('Product_Category_1').mean()['Purchase'].sort_values().plot(kind = 'bar')
```

# ->Which products have been the most amount spent on ?

```python
df.groupby('Product_ID').sum()['Purchase'].nlargest(10).sort_values().plot(kind = 'bar')
```

# ->The most selling Products

```python
df.groupby('Product_ID').size().nlargest(10).sort_values().plot(kind = 'bar')
```

# ->Most costly products

```python
df.groupby('Product_ID').mean()['Purchase'].nlargest(10).sort_values().plot(kind = 'bar')
```
