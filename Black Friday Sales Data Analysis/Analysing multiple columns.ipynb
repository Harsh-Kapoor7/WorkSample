```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv('D:/Data_Analysis/BlackFriday.csv')

del df['Product_Category_2']
del df['Product_Category_3']
```

```python
df.head()
```

```python
df.groupby('Age').size().plot(kind = 'bar', figsize = (10,4))
```

```python
sns.set(rc = {'figure.figsize' : (8,4)})
sns.countplot(x = df['Age'], hue = df['Gender'])
```

```python
sns.set(rc = {'figure.figsize' : (8,4)})
sns.countplot(x = df['Age'] == '55+', hue = df['Gender'])
```

```python
sns.set(rc = {'figure.figsize' : (7,4)})
sns.countplot(x = df['Gender'], hue = df['Marital_Status'])
```
