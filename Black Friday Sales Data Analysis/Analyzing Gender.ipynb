```python
df.groupby('Gender').size().plot(kind = 'pie',
                                 autopct = '%.2f',
                                 figsize = (5,5),
                                 title = 'Gender Ratio')
```

```python
df.groupby('Gender').size().plot(kind = 'bar',
                                 figsize = (5,5),
                                 title = 'Gender Ratio')
```

```python
df.groupby('Gender').sum()['Purchase'].plot(kind = 'pie', autopct = '%.2f', title = 'Purchase Ratio')
```

```python
df_pr = df.groupby('Gender').sum()['Purchase']

df_pr.plot(kind = 'bar', title = 'Purchase Ratio')
```

```python
df.groupby('Gender').mean()['Purchase'].plot(kind = 'pie', autopct = '%.2f', title = 'Average Purchase Ratio')
```
